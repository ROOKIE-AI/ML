# 什么是机器学习?

> [课程:什么是机器学习?](https://www.bilibili.com/video/BV1Bq421A74G/?p=3)

## 个人理解

&emsp;&emsp;机器学习是一种通过从数据中学习和自动改进的技术。通俗地说，机器学习就像是教电脑从经验中学习，让它能够在没有明确编程指令的情况下完成任务。

&emsp;&emsp;假设你有一只新宠物狗，你想教它坐下。传统编程就像是你详细地告诉狗每一步应该怎么做，比如：“首先弯下前腿，然后把身体降到地面。” 这就像是给电脑写程序，每一步都明确指令。而机器学习就像是你通过示范和奖励的方法教狗学会坐下。你不断地给它示范“坐下”，然后当它正确地模仿你时，你奖励它一个小零食。经过多次重复，它就会学会“坐下”这个动作，即使你不再示范或指令，它也能理解和执行这个动作。

&emsp;&emsp;一直以来，人类对普遍性问题的认识通常有两种途径：基于经验进行归纳和通过理论进行演绎推理。<br>
&emsp;&emsp;所谓经验，是指通过前人不断的实践，总结归纳出来的一些普适性法则。比如，古代人类根据观察天象、气象变化，形成了许多天气预测的智慧谚语，如“红霞夕阳，明日晴朗”，这些谚语是通过长期观察总结而成，具有一定的预测准确性。又如《道德经》、《论语》等普世性论著，蕴含着丰富的人生智慧，是历代智者在实践中的经验总结，对人类思维与行为具有深刻启示。这种认识事物的方法被称为归纳法。<br>
&emsp;&emsp;另一种途径是通过现代科学理论，从具体的理论进行一般性推演，基于第一性原理进行思考。例如，牛顿的万有引力定律，通过观察苹果落地、行星运动等具体现象，牛顿推导出了普适的引力定律，从而解释了天体运动的规律。这一理论应用在航天工程中，使得人类能够准确地预测和规划航天器的轨道和运行轨迹，实现了人类探索宇宙的壮举。再如爱因斯坦的相对论，从对光速不变的假设出发，推演出了时间、空间的相对性，为现代物理学的发展开辟了新的领域。这一理论的应用在 GPS 定位系统中起到了关键作用，通过修正时间差异，确保了卫星定位系统的准确性，为我们提供了精准的导航服务。这种方法通过科学的逻辑和原理，从特定的理论推导出广泛适用的结论，帮助我们理解和预测复杂的现象。

&emsp;&emsp;
然而，由于人类的科学发现尚不足以解释所有现象，同时对于某些人类直觉性感知理解方面的问题，尚无法通过明确的步骤进行归纳和演绎。因此，基于第一性原理的科学理论还不足以解决人类在现实世界中感知、预测、推理等方面的全部问题。人类感知系统和大脑本身的物理局限性，决定了人类对很多问题并不擅长解决，自然很难从中总结出普适性的规律用于演绎推理。因此，通过演绎和归纳 解释现实中更多的一般性问题依然是一项挑战。
