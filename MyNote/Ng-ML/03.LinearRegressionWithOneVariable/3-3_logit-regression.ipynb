# 对数几率回归

&emsp;&emsp;之前讨论使用线性模型进行回归学习, 但是如何用它来实现分类任务呢？<br>
&emsp;&emsp;答案蕴含在广义线性模型中:只需要找到一个单调且可微的函数将分类任务的真实标记y和线性模型的预测值联系起来。

> 方程如下:<br>
> &emsp;&emsp; $y = \frac{1}{1+e^{-(w^Tx+b)}}$<br><br>
> &emsp;&emsp;==> &emsp;$ln\frac{y}{1-y} = w^Tx + b$ 

&emsp;&emsp;&emsp;&emsp;$ sigmid(x) = \frac{1}{1+e^z}$

```python
import numpy as np
import matplotlib.pyplot as plt
# 定义 Sigmoid 函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.linspace(-10, 10, 100)
# 计算 Sigmoid 函数值
y = sigmoid(x)
# 绘制 Sigmoid 函数
fig, ax = plt.subplots()
ax.plot(x, y)
ax.spines['left'].set_position(('data', 0))
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['bottom'].set_color('black')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
plt.xlabel('x')
plt.title('Sigmoid Function')
plt.show()
```

> Sigmoid 函数是非线性函数，它的形状是非凸的。
