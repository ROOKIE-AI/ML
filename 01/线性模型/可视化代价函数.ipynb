```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.sort((np.random.random(100) * 100))
```

```python
def linear_func(x, w, b = 0.0):
    return x * w + b + np.random.normal(0, 2, size=x.shape)
```

```python
y = linear_func(x, 2.0)
```

```python
plt.scatter(x, y, s=5)
plt.show()
```

```python
def loss_func(x, y, w):
    return 1/2 * ((x * w  - y) ** 2).mean()
ws = np.linspace(0, 4, 100)
jw = [loss_func(x, y, i) for i in ws]
plt.plot(ws, jw)
plt.show()
```

```python
y = linear_func(x, 2.0, -40.2)
```

```python
plt.scatter(x, y, s=5)
plt.show()
```

```python
def loss_func(x, y, w, b):
    return 1/2 * ((x * w + b  - y) ** 2).mean()
ws = np.linspace(0, 4, 100)
bs = np.linspace(-50, 0, 10000)
jw = np.array([[loss_func(x, y, w, b) for w in ws] for b in bs])
jw.shape, bs[np.argmin(jw) // 100], ws[np.argmin(jw) % 100]
```

```python
W, B = np.meshgrid(ws, bs)
plt.contourf(W, B, jw, levels=100, cmap='viridis')
plt.colorbar()  # 添加颜色条，帮助理解色阶
plt.title('Loss Function Contour')
plt.xlabel('Weight (w)')
plt.ylabel('Bias (b)')
plt.show()
```

## 最小二乘法

```python
w = ((x - x.mean()) * y ).sum() / ((x**2).sum() - x.sum()**2/x.shape[0])
b = (y - w * x).mean()
w, b
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# 获取回归系数（斜率 w）和截距（b）
w = model.coef_[0]  # 斜率
b = model.intercept_  # 截距

w, b
```
