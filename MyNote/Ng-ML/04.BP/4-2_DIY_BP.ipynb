# 动手实现基于一元线性函数梯度下降

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
true_w, true_b = 2.1, -3.8
x = np.linspace(0, 10, 100)
y = true_w * x + true_b
```

```python
def calc_y_hat(x, w, b):
    return x * w + b

def GradientDescent(x, y, w, b,  alpha, epochs):
    for i in range(epochs):
        y_hat = calc_y_hat(x, w, b)
        J = ((y - y_hat)**2).mean()
        # print(y.shape, x.shape)
        Jw = -2 * ((y - y_hat) * x).mean()
        Jb = -2 * ((y - y_hat)).mean()
        # print(Jw.shape)
        w =  w - alpha * Jw
        b = b - alpha * Jb
        print(f"epochs :{i}, Loss={J}. Jw = {Jw}, Jb = {Jb}")

    return w, b
```

```python
init_w, init_b = 1.0, 1.0
alpha = 0.01
epochs = 10000
```

```python
w, b = GradientDescent(x, y, init_w, init_b, alpha, epochs)
```
