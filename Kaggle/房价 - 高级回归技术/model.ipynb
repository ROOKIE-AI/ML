## Feature Engineering

### 1. one-hat

```python
df = pd.concat([train_df, test_df])
df.head()
```

```python
one_hat_featue_names = df.columns[(df.dtypes == np.dtype(np.object_)).values] # type: ignore
for name in one_hat_featue_names:
    categories = df[name].unique()
    df = pd.get_dummies(df, columns=[name], prefix=name)
```

```python
pprint(df.columns.values)
```

### 2. Missing value handling

```python
missing_rate = df.isnull().mean().loc[lambda x: x > 0] * 100
missing_rate
```

```python
# 直接剔除缺失比例超过60%的因子
train_df.drop(list(missing_rate[missing_rate > 60].index.values), axis=1, inplace=True)
```

```python
# 对缺失比例低于60%的因子选择用中位数替代
for name in missing_rate[missing_rate < 60].index.values:
    if name == 'SalePrice':
        continue
    df[name].fillna(df[name].median(), inplace=True)
```

```python
df.isnull().mean().loc[lambda x:  x > 0]
```

```python
df.shape
```

### 3. feature scaling

```python
_X_train, _y_train = df[df['Id'].isin(train_df['Id'])].drop('SalePrice', axis=1).values, df[df['Id'].isin(train_df['Id'])]['SalePrice'].values.reshape(-1, 1)
X_test = df[df['Id'].isin(test_df['Id'])].drop('SalePrice', axis=1).values
```

```python
_y_train.min()
```

```python
_X_train.shape, _y_train.shape, X_test.shape
```

```python
scaler_X = MinMaxScaler()
scaler_y = StandardScaler()
X_train_scaler = scaler_X.fit_transform(_X_train)
X_test_scaler = scaler_X.transform(X_test)
y_train_scaler = scaler_y.fit_transform(_y_train)
```

## Modeling

```python
X_train, X_valid, y_train, y_vaild = train_test_split(X_train_scaler, y_train_scaler, test_size=0.1, random_state=42)
y_vaild = scaler_y.inverse_transform(y_vaild)
```

```python
y_vaild.min()
```

```python
models = {
    'LR': LinearRegression(),
    'RF': RandomForestRegressor(n_estimators=100),
    'XGB': XGBRegressor(),
    'MLP': MLPRegressor(hidden_layer_sizes=(400, 100, 10), max_iter=300)
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    y_pred = scaler_y.inverse_transform(y_pred)
    # print(y_pred.min())
    y_pred = np.where(y_pred <= 0, y_vaild.min(), y_pred)
    msle = mean_squared_log_error(y_pred, y_vaild)
    print(f"{model_name}: {msle}")
```

## Submit

```python
y_test_pred_scaler = models['XGB'].predict(X_test_scaler)
y_test_pred = scaler_y.inverse_transform(y_test_pred_scaler)
y_test_pred.shape
```

```python
output = pd.DataFrame({'Id': test_df.Id, 'SalePrice': y_test_pred})
output.to_csv('./data/submission.csv', index=False)
print("Your submission was successfully saved!")
```
